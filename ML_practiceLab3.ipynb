```python
import pandas as pd
dataframe = pd.read_csv("phone_data.csv")

dataframe.drop('id', inplace=True, axis=1)
print(dataframe)
```

```python
print("Total features :", len(dataframe.columns) -1)
print("Total patterns :", dataframe.shape[0])


minimum = dataframe['price_range'].min()
maximum = dataframe['price_range'].max()
print("Range of output :", (minimum, maximum))
```

```python
x = .1
while x <= .9:
  y=1-x
  dataset1 = dataframe.copy()
  x_part = dataset1.sample(frac = x)
  #print(x_part.index)
  y_part = dataset1.drop(x_part.index)

  print("Showing the dataset ....", x, ':' , y)
  print("x part ...")
  print(x_part)
  print("y part ...")
  print(y_part)
  x+=0.1
```

```python
dataset = dataframe
print("Row wise mean ....\n", dataset.mean(axis=1))
print("\n\n\nrow wise median....\n", dataset.median(axis=1))
print("\n\n\nrow wise mode....\n", dataset.mode(axis=1))
print("\n\n\nrow wise standard deviation....\n", dataset.std(axis=1))
```

```python
print("column wise mean ....\n", dataset.mean(axis=0))
print("\n\n\ncolumn wise median....\n", dataset.median(axis=0))
print("\n\n\ncolumn wise mode....\n", dataset.mode(axis=0))
print("\n\n\ncolumn wise standard deviation....\n", dataset.std(axis=0))
```

```python
print("Overall mean ....\n", dataset.mean())
print("\n\n\nOverall median....\n", dataset.median())
print("\n\n\nOverall mode....\n", dataset.mode())
print("\n\n\nOverall standard deviation....\n", dataset.std())
```
